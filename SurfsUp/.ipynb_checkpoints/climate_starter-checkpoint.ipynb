# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Hawaii_m = Base.classes.measurement
Hawaii_s= Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
recent_date = session.query(Hawaii_m.date).order_by(Hawaii_m.date.desc()).first()
recent_date
```

```python
# Find the most recent date in the data set.
for date in recent_date: 
    latest_date = pd.to_datetime(date)
```

```python
latest_date.year,latest_date.month,latest_date.day
```

```python
#add the date using DT
```

```python
date_year = dt.date(latest_date.year,latest_date.month,latest_date.day)-dt.timedelta(days = 365)
date_year
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
# Calculate the date one year from the last date in data set.
# Perform a query to retrieve the data and precipitation scores
rain_by_year = []
precipitation_12 = session.query(Hawaii_m.date,Hawaii_m.prcp).filter(Hawaii_m.date >= date_year).all()
for rain in precipitation_12:
    rain_by_year.append(rain)

# Save the query results as a Pandas DataFrame. Explicitly set the column names
rain_df = pd.DataFrame(rain_by_year)
rain_df.set_index('date',inplace=True)


# Sort the dataframe by date
rain_df.sort_values(by = "date", inplace = True)

# Use Pandas Plotting with Matplotlib to plot the data
plt.rcParams['figure.figsize']=(9,7)
rain_df.plot(linewidth=2,alpha=1,rot=90)
plt.xlabel('Date',fontsize=16,color='black',labelpad=20)
plt.ylabel('Precipitation (in)',fontsize=16,color='black',labelpad=20)
plt.title('Daily Precipitation for One Year',fontsize=15,pad=40)

plt.show()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
rain_df.describe()
```

# Exploratory Station Analysis

```python
data = pd.read_sql("SELECT * FROM measurement", engine)
data
```

```python
# Design a query to calculate the total number of stations in the dataset
station_count = session.query(func.count(Hawaii_s.station)).first()
station_count
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
station_activity= session.query(Hawaii_m.station,func.count(Hawaii_m.station)).\
group_by(Hawaii_m.station).\
order_by(func.count(Hawaii_m.station).desc())

for station in station_activity:
    print(station)
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
temp=[       func.min(Hawaii_m.tobs), 
             func.max(Hawaii_m.tobs), 
             func.avg(Hawaii_m.tobs)]

most_active_st=(session.query(*temp)
                       .filter(Hawaii_m.station=='USC00519281')).first()
                       
most_active_st
```

```python
#filter the data by the most active station and activity from one year ago to most recent. 
frequency_year = []
active_st_year = session.query(Hawaii_m.tobs,Hawaii_m.date).filter(Hawaii_m.station=='USC00519281').\
                filter(Hawaii_m.date >= date_year)


for year in active_st_year:
    frequency_year.append(year)
frequency_data = pd.DataFrame(frequency_year)
```

```python
frequency_data.set_index('date',inplace=True)
```

```python
frequency_data.sort_values(by = "date", inplace = True)
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
plt.rcParams['figure.figsize']=(10,7)
plt.hist(frequency_data['tobs'],bins=12,alpha=0.6,linewidth=1,
         edgecolor='xkcd:gray',color='xkcd:blue' )


plt.xlabel('Temperature',fontsize=15,color='black',labelpad=20)
plt.ylabel('Frequency',fontsize=15,color='black',labelpad=20)
plt.title('Temperature Frequency',fontsize=15,pad=40)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(0,70)

plt.show()
```
